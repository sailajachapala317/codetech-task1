```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('iris')
print("Dataset Information:")
print(df.info())
print("\n")
print("Descriptive Statistics:")
print(df.describe())
print("\n")
print("Null Values:")
print(df.isnull().sum())
print("\n")
plt.figure(figsize=(12, 8))
df.hist(bins=20, edgecolor='black', figsize=(12, 8))
plt.suptitle('Feature Distributions', fontsize=16)
plt.show()
sns.pairplot(df, hue='species', palette='Set1')
plt.suptitle('Pairwise Relationships', fontsize=16)
plt.show()
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, width=0.5, palette='Set1')
plt.title('Boxplots for Features', fontsize=16)
plt.show()
plt.figure(figsize=(10, 8))
correlation_matrix = df.drop('species', axis=1).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap', fontsize=16)
plt.show()
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df, palette='Set1')
plt.title('Sepal Length vs Sepal Width', fontsize=16)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
plt.figure(figsize=(12, 8))
sns.violinplot(x='species', y='sepal_length', data=df, palette='Set1')
plt.title('Violin Plot for Sepal Length by Species', fontsize=16)
plt.show()
sns.pairplot(df, hue='species', kind='reg', palette='Set1')
plt.suptitle('Pairwise Scatter Plot with Regression Lines', fontsize=16)
plt.show()
```
